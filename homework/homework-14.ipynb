### Sample Solution 14.3

```python
#@title Sample Solution 13.9 {display-mode: "form"}
dot = '''
digraph {
    rankdir=LR;
    0 [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; A;
    node [shape=circle, fontsize=8, margin=0]; REJECT;
    node [shape=doublecircle, margin=0, fontsize=8]; ACCEPT;
    0->A [color=red];
    A->A [label=0];
    A->ACCEPT [label=1];
    ACCEPT->REJECT [label="0,1"];
    REJECT->REJECT [label="0,1"];
}
'''
from graphviz import Source
Source(dot)
```

```python
#@title Alternative Sample Solution 13.9 {display-mode: "form"}
dot = '''
digraph {
    rankdir=LR;
    0 [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; A B;
    node [shape=circle, fontsize=8, margin=0]; REJECT;
    node [shape=doublecircle, margin=0, fontsize=8]; ACCEPT;
    0->A [color=red];
    A->B [label=0];
    B->B [label=0];
    B->ACCEPT [label=1];
    A->ACCEPT [label=1];
    ACCEPT->REJECT [label="0,1"];
    REJECT->REJECT [label="0,1"];
}
'''
from graphviz import Source
Source(dot)
```

### Sample Solution 14.4

```python
#@title Sample Solution 13.10 {display-mode: "form"}

# Recognized bitstrings whose third-to-last bit is a 0
machine = '''
digraph {
    layout=dot;
    rankdir=LR;
    ranksep=1;
    size=6;
    start [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    subgraph {
        node [shape=circle]; s 0 00 01;
    }
    subgraph {
        rank=sink;
        node [shape=doublecircle]; 000 001 010 011;
    }
    start->s [color=red];
    s->s [label=1];
    s->0 [label=0];
    0->00 [label=0];
    0->01 [label=1];
    00->000 [label=0];
    00->001 [label=1];
    01->010 [label=0];
    01->011 [label=1];
    000->000 [label=0];
    000->001 [label=1];
    001->010 [label=0];
    001->011 [label=1, labelfloat="true"];
    010->00 [label=0];
    010->01 [label=1];
    011->0 [label=0];
    011->s [label=1];
}
'''
from graphviz import Source
Source(machine)
```
