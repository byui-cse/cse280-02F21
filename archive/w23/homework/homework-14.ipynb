### Exercise 14.4


The following graph shows a machine that recognizes the language of all bitstrings whose *second-to-last* bit is a `0`. Using this as a pattern, create (and draw) a recognizer of bitstrings whose *third-to-last* bit is a `0`.

Each state, including the accept states, should have a transition for 0 and for 1.

```python
#@title #### FSM {display-mode: "form"}

# Code to draw graph for exercise 14.4
machine = '''
digraph {
    rankdir=LR;
    start [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; s0 s1;
    subgraph {
        rank=same;
        node [shape=doublecircle]; s2 s3;
    }
    start->s0 [color=red];
    s0->s0 [label=1];
    s0->s1 [label=0];
    s1->s2 [label=1];
    s1->s3 [label=0];
    s2->s1 [label=0];
    s2->s0 [label=1];
    s3->s3 [label=0];
    s3->s2 [label=1];
}
'''
from graphviz import Source
from IPython.display import Image
display(Image(Source(machine, format='png').render()))
```

Hint: You'll need an accept state for each of the four possible ending patterns of three bits, i.e., 000, 001, 010, 011. One way to help you visualize this is to label each state with the pattern of bits required to reach that state. For example, we could re-label the first FSM like this:

```python
#@title #### Re-labeled FSM {display-mode: "form"}

# Code to draw graph for exercise 14.4
machine = '''
digraph {
    rankdir=LR;
    start [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; s 0;
    subgraph {
        rank=same;
        node [shape=doublecircle]; 01 00;
    }
    start->s [color=red];
    s->s [label=1];
    s->0 [label=0];
    0->01 [label=1];
    0->00 [label=0];
    01->0 [label=0];
    01->s [label=1];
    00->00 [label=0];
    00->01 [label=1];
}
'''
from graphviz import Source
from IPython.display import Image
display(Image(Source(machine, format='png').render()))
```

### Sample Solution 14.3

```python
#@title Sample Solution 14.3 {display-mode: "form"}
dot = '''
digraph {
    rankdir=LR;
    0 [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; A;
    node [shape=circle, fontsize=8, margin=0]; REJECT;
    node [shape=doublecircle, margin=0, fontsize=8]; ACCEPT;
    0->A [color=red];
    A->A [label=0];
    A->ACCEPT [label=1];
    ACCEPT->REJECT [label="0,1"];
    REJECT->REJECT [label="0,1"];
}
'''
from graphviz import Source
from IPython.display import Image
display(Image(Source(dot, format='png').render()))
```

```python
#@title Alternative Sample Solution 14.3 {display-mode: "form"}
dot = '''
digraph {
    rankdir=LR;
    0 [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    node [shape=circle]; A B;
    node [shape=circle, fontsize=8, margin=0]; REJECT;
    node [shape=doublecircle, margin=0, fontsize=8]; ACCEPT;
    0->A [color=red];
    A->B [label=0];
    B->B [label=0];
    B->ACCEPT [label=1];
    A->ACCEPT [label=1];
    ACCEPT->REJECT [label="0,1"];
    REJECT->REJECT [label="0,1"];
}
'''
from graphviz import Source
from IPython.display import Image
display(Image(Source(dot, format='png').render()))
```

### Sample Solution 14.4

```python
#@title Sample Solution 14.4 {display-mode: "form"}

# Recognized bitstrings whose third-to-last bit is a 0
machine = '''
digraph {
    layout=dot;
    rankdir=LR;
    ranksep=1;
    size=6;
    start [shape=rect, label="Start", color="red", fontcolor="red", fontsize=12, width=0.5, height=0.25];
    subgraph {
        node [shape=circle]; s 0 00 01;
    }
    subgraph {
        rank=sink;
        node [shape=doublecircle]; 000 001 010 011;
    }
    start->s [color=red];
    s->s [label=1];
    s->0 [label=0];
    0->00 [label=0];
    0->01 [label=1];
    00->000 [label=0];
    00->001 [label=1];
    01->010 [label=0];
    01->011 [label=1];
    000->000 [label=0];
    000->001 [label=1];
    001->010 [label=0];
    001:e->011 [taillabel=1, labeldistance=5, constraint=false];
    010->00 [label=0];
    010->01 [label=1];
    011->0 [label=0];
    011->s [label=1];
}
'''
from graphviz import Source
from IPython.display import Image
display(Image(Source(machine, format='png').render()))
```
