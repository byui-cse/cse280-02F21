## <font color=red>Teaching Combinatorics through Examples</font>

Do small cases first.

### <font color=red>Committees</font>


From 4 departments, must make size-3 committees, so that no 2 people in a committee are from the same department.


|   | Dept | # of people |
|---|------|------------:|
| 1 | CSE  |          30 |
| 2 | ECE  |          35 |
| 3 | Math |          25 |
| 4 | Bio  |          40 |


How many possible committees could be made?


What to do?


Break the problem up into smaller parts then add them together using the Sum Rule.


Use a template:


| Depts | 1 | 2 | 3 | 4 | Product Rule |        |
|-------|---|---|---|---|--------------|-------:|
|       | X | X | X | - | 30 * 35 * 25 |  26250 |
|       | X | X | - | X | 30 * 35 * 40 |  42000 |
|       | X | - | X | X | 30 * 25 * 40 |  30000 |
|       | - | X | X | X | 35 * 25 * 40 |  35000 |
|       |   |   |   |   | Total        | 133250 |


Can be more complicated: what if we want "gender balance"?


### <font color=red>Sandwiches</font>


| Set | Sandwich component | How many kinds? |
|-----|--------|--:|
|  B  | bread  | 3 |
|  C  | cheese | 4 |
|  S  | spread | 5 |
|  M  | meat   | 6 |


How many different kinds of sandwiches are there?


$X = \{(b, c, s, m)\ \vert\ b \in B, c \in C, s \in S, m \in M\}$


$\vert X \vert = $?


$X = R_0 \cup R_1 \cup R_2 \cup R_3 \cup R_4 \cup R_5 \cup R_6 \cup R_7$.


$\vert X \vert = \vert R_0 \vert + \vert R_1 \vert + \vert R_2 \vert + \vert R_3 \vert + \vert R_4 \vert + \vert R_5 \vert + \vert R_6 \vert + \vert R_7 \vert$

## <font color=red>Another Counting/Probability Problem</font>

(Count only rectangles with a non-zero area, and with horizontal and vertical sides, not ones with other slanted angles.)
