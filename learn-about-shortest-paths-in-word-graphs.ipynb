## <font color=red>Definition</font>


Word graphs are like those in Exercise 8.11.3.


The nodes are words, the links are between words that differ by only one letter in the same position (e.g., **wind** and **wins** differ in the fourth letter only, so that defines a link. But **dine** and **tone** differ in the first and second letters, so no link there).


````python
# Is there a better way to tell?
def link_exists(w1, w2):
  return len(w1) == len(w2) and len([i for i in range(len(w1)) if w1[i] != w2[i]]) == 1
````
