## Truth tables in Python

Generate the truth table for each of the following compound propositions using Python.

$p \land q$

$p \lor q$

$p \oplus q$

$p \rightarrow q$

$\neg p \wedge (q \vee r)$


### $p \land q$

```python
# Truth table for p and q

print("p  q  p and q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}  {p and q}")
```

#### With True/False

```python
# Truth table for p and q using True/False and nicely formatted
def p_and_q():
  print(f"{'p':>10}{'q':>10}{'p and q':>10}")
  for p in [True, False]:
    for q in [True, False]:
      print(f"{p!s:>10}{q!s:>10}{p and q!s:>10}") # print True/False instead of 1/0
      # print(f"{p:>10}{q:>10}{p and q:>10}")

p_and_q()
```

### $p \lor q$

```python
# Truth table for p or q

print("p  q  p or q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}  {p or q}")

```

### $p \oplus q$

```python
# Truth table for p xor q

print("p  q  p xor q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}  {p ^ q}")

```

### $p \implies q$

```python
# Truth table for p -> q

print("p  q  p -> q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}  {not p or q:d}")

```

### $\neg p \wedge (q \vee r)$

```python
# Truth table for not p and (q or r)

print("p  q  not p and (q or r)")
for p in [0,1]:
  for q in [0,1]:
    for r in [0,1]:
      print(f"{p}  {q}  {not p and (q or r):d}")

```

## Implications (Conditionals)


### Implement in python

$p \implies q$

```python
# implement the conditional statement using python
def implies(p, q):
  if p:
    return q
  else:
    return True


# verify this implementation is correct by building a truth table
print("p  q    p -> q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}    {implies(p,q):d}")
```

#### Another way

```python
# Another way
def implies_v2(p, q):
  return not p or q

# verify this implementation is correct by building a truth table
print("p  q    p -> q")
for p in [0,1]:
  for q in [0,1]:
    print(f"{p}  {q}    {implies_v2(p,q):d}")
```

### Practical usage

Is this implication true for all x and y?

$x \ge 0 \wedge y \ge 0 \implies x*y \ge 0$


```python
for x in range(-5,5):
  for y in range(-5,5):
    print(implies(x>=0 and y>=0, x*y>=0))
```
