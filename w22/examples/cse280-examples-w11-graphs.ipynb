Using Graphviz

http://www.graphviz.org/

```python
dot = '''
graph { 
  layout=neato
  node [shape=circle]
  1; 2; 3; 4; 5;
  1 -- 2;
  1 -- 3;
  2 -- 4;
  2 -- 5;
  3 -- 5;
}'''
from graphviz import Source
Source(dot)
```

What is the degree sequence of this graph?



[3,2,2,2,1]

Definition of a graph:


```python
dot = '''
graph { 
  layout=neato
  node [shape=circle]
  A; B; C; D;
  A -- B;
  A -- C;
  B -- C;
  C -- D;
}'''
from graphviz import Source
Source(dot)
```

```python
dot = '''
graph {
    layout=neato;
    size="2,2";
    node [shape=circle, height=.1];

    a; b; c; d; e; f;

    a -- b;
    a -- c;
    b -- c;
    b -- f;
    b -- e;
    c -- f;


  }
 '''
from graphviz import Source
Source(dot)
```

```python
dot = '''
graph {
    layout=neato;
    node [shape=circle];

    a; b; c; d; e;

    a -- b [color=green];
    a -- c;
    a -- d [color=green];
    a -- e;
    b -- c [color=green];
    b -- d [color=green];
    b -- e;
    c -- d;
    c -- e [color=green];
    d -- e;
  }
 '''
from graphviz import Source
Source(dot)
```
