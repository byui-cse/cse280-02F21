## <font color=red>Exercise 1.4.1</font>

Searching the web based on the hint and the one word that did not get dehydrated, using the search terms *Carroll Cabbages* reveals the answer. This is from the nonsense poem Carroll called **The Walrus and the Carpenter** in his 1872 book *Through the Looking-Glass and What Alice Found There*:

```THE TIME HAS COME THE WALRUS SAID TO TALK OF MANY THINGS OF SHOES AND SHIPS AND SEALING WAX OF CABBAGES AND KINGS AND WHY THE SEA IS BOILING HOT AND WHETHER PIGS HAVE WINGS```

A possible solution that could be created for verifying the textual transformation:

````python
def dehydrate(text):
  result = ''
  for c in text:
    result += ('' if c in 'HIJKLMNO' else c)
  return result
````

````python
text = '''
THE TIME HAS COME THE WALRUS SAID TO TALK OF MANY THINGS OF SHOES AND SHIPS
AND SEALING WAX OF CABBAGES AND KINGS AND WHY THE SEA IS BOILING HOT
AND WHETHER PIGS HAVE WINGS
'''
print(dehydrate(text))
````

## <font color=red>Exercise 1.4.2</font>

One familiar word is 'strengths' with a CVR of 8 to 1. If you look you can find other less common words with a higher CVR.

The following code is not quite correct. What is the incorrect assumption it makes, and how would you fix it?

````python
def CVR(word):
  word = word.upper()
  num_vowels = ('A' in word) + ('E' in word) + ('I' in word) + ('O' in word) + ('U' in word)
  num_consonants = len(word) - num_vowels
  return num_consonants / num_vowels
````

````python
CVR('strengths')
````

## <font color=red>Exercise 1.4.3</font>

The numbers one, five, fourteen, thirty, fifty-five, etc.are called the square pyramidal numbers.


````python
def nth_square_pyramidal_number(n):
  return (n * (n + 1) * (2 * n + 1)) // 6
````

````python
for n in range(1, 6):
  print(nth_square_pyramidal_number(n))
````

Indexing at 0, consonants number 9 (M), 7 (K), 5 (H), 3 (F), and 1 (C) are interleaved with all five vowels: 0 (A), 1 (E), 2 (I), 3 (O) and 4 (U).

Solution by Kyle Mueller:

````python
vowels = ['A','E','I','O','U']
consonants = ['B','C','D','F','G','H','J','K','L','M','N','P','Q','R','S','T','V','W','X','Y','Z']

c2n = lambda c:[consonants,vowels][c in vowels].index(c)

[*map(c2n,"MAKEHIFOCU")]
````

## <font color=red>Exercise 1.4.5</font>

The limiting radius ($r_\infty$) of the outermost circle is 8.700 (approximately).

A possible solution that could be created for finding this limiting radius:

````python
from math import cos, pi

radius = 1.0

for n in range(3, 10000000):
  radius *= 1 / cos(pi / n)

print(radius)
````
